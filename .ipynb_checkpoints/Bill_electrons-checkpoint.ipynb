```python
from ROOT import TF2, TH1D, TF1, TFile, TCanvas
import matplotlib.pyplot as plt
from ROOT import gROOT 
import numpy as np
import ROOT
import pandas as pd

# Sample a ROOT function
# myfile1 = TFile( 'simphotons_Geo_v01.root', 'READ' );
# myfile2 = TFile( 'simphotons_Geo_v02.root', 'READ' );

file_10mev=TFile( 'bill_electrons_10MeV.root', 'READ' );
file_100mev=TFile( 'bill_electrons_100MeV.root', 'READ' );
file_1000mev=TFile( 'bill_electrons_1000MeV.root', 'READ' );

ntupla_xarapuca_10mev=file_10mev.Get("myanalysis/xarapuca");
ntupla_xarapuca_100mev=file_100mev.Get("myanalysis/xarapuca");
ntupla_xarapuca_1000mev=file_1000mev.Get("myanalysis/xarapuca");

ntupla_IDE_10mev=file_10mev.Get("myanalysis/IDEs");
ntupla_IDE_100mev=file_100mev.Get("myanalysis/IDEs");
ntupla_IDE_1000mev=file_1000mev.Get("myanalysis/IDEs");

array_ntupla_xarapuca_10mev=ntupla_xarapuca_10mev.AsMatrix();
array_ntupla_xarapuca_100mev=ntupla_xarapuca_100mev.AsMatrix();
array_ntupla_xarapuca_1000mev=ntupla_xarapuca_1000mev.AsMatrix();

array_ntupla_IDE_10mev=ntupla_IDE_10mev.AsMatrix();
array_ntupla_IDE_100mev=ntupla_IDE_100mev.AsMatrix();
array_ntupla_IDE_1000mev=ntupla_IDE_1000mev.AsMatrix();


# ntupla_xarapuca.ls()
```

```python
#python approach
# plt.hist(array_ntupla_xarapuca_10mev[:,1],320,[1, 320]);
# plt.title("X-Arapucas: Incoming Photons per Channel")
names=["ev", "totalDE","totalPEarapucas","totalPEarapucasVUV","totalPEarapucasVis"]

df_10MeV = pd.DataFrame(data=array_ntupla_IDE_10mev, columns=names)
df_100MeV = pd.DataFrame(data=array_ntupla_IDE_100mev, columns=names)
df_1000MeV = pd.DataFrame(data=array_ntupla_IDE_1000mev, columns=names)

print (df_1000MeV)
```

```python
PE_per_MeV_10MeV=df_10MeV[:]["totalPEarapucas"]/df_10MeV[:]["totalDE"];
PE_per_MeV_100MeV=df_100MeV[:]["totalPEarapucas"]/df_100MeV[:]["totalDE"];
PE_per_MeV_1000MeV=df_1000MeV[:]["totalPEarapucas"]/df_1000MeV[:]["totalDE"];

VUVPE_per_MeV_10MeV=df_10MeV[:]["totalPEarapucasVUV"]/df_10MeV[:]["totalDE"];
VUVPE_per_MeV_100MeV=df_100MeV[:]["totalPEarapucasVUV"]/df_100MeV[:]["totalDE"];
VUVPE_per_MeV_1000MeV=df_1000MeV[:]["totalPEarapucasVUV"]/df_1000MeV[:]["totalDE"];

VisPE_per_MeV_10MeV=df_10MeV[:]["totalPEarapucasVis"]/df_10MeV[:]["totalDE"];
VisPE_per_MeV_100MeV=df_100MeV[:]["totalPEarapucasVis"]/df_100MeV[:]["totalDE"];
VisPE_per_MeV_1000MeV=df_1000MeV[:]["totalPEarapucasVis"]/df_1000MeV[:]["totalDE"];

fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
ax = fig.add_subplot()

# plot settings
hist_type="bar"
line_width=1
alpha_=1
nbins=80


plt.hist(PE_per_MeV_10MeV,nbins,[3,10],density=True,histtype=hist_type,alpha=alpha_,linewidth=line_width);
plt.hist(PE_per_MeV_100MeV,nbins,[3,10],density=True,histtype=hist_type,alpha=alpha_,linewidth=line_width);
plt.hist(PE_per_MeV_1000MeV,nbins,[3,10],density=True,histtype=hist_type,alpha=alpha_,linewidth=line_width);

plt.axvline(PE_per_MeV_10MeV.mean(),color="blue",linestyle='dashed')
plt.axvline(PE_per_MeV_100MeV.mean(),color="orange",linestyle='dashed')
plt.axvline(PE_per_MeV_1000MeV.mean(),color="green",linestyle='dashed')
lg=["$10$ MeV: <PE/MeV>=%.2f" %PE_per_MeV_10MeV.mean(),"$100$ MeV: <PE/MeV>=%.2f" %PE_per_MeV_100MeV.mean(),"$1 GeV$ MeV: <PE/MeV>=%.2f" %PE_per_MeV_1000MeV.mean()]
plt.legend(lg,fontsize=15)
plt.title("$e^-$ 1.85m from Anode: All X-Arapuca Channels")
plt.gca().set_xlabel("Arriving PE/MeV",fontsize=20)
plt.gca().set_ylabel("Normalized Counts",fontsize=20)

```
